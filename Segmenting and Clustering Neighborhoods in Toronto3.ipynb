# Explore Neighborhoods in Toronto

#### Now, let's get the top 100 venues that are within a radius of 500 meters.

```python
import requests
radius = 500
LIMIT = 100

venues = []

for lat, long, post, b, n in zip(df3['Latitude'],df3['Longitude'],df3['Postal Code'],df3['Borough'],df3['Neighborhood']):
    url = "https://api.foursquare.com/v2/venues/explore?client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}".format(
        CLIENT_ID,
        CLIENT_SECRET,
        VERSION,
        lat,
        long,
        radius, 
        LIMIT)
    
    results = requests.get(url).json()["response"]['groups'][0]['items']
    
    for venue in results:
        venues.append((
            post, 
            b,
            n,
            lat, 
            long, 
            venue['venue']['name'], 
            venue['venue']['location']['lat'], 
            venue['venue']['location']['lng'],  
            venue['venue']['categories'][0]['name']))
```

```python
vdf = pd.DataFrame(venues)

# define the column names
vdf.columns = ['Postal Code', 'Borough', 'Neighborhood', 'BoroughLatitude', 'BoroughLongitude', 'VenueName', 'VenueLatitude', 'VenueLongitude', 'VenueCategory']
vdf
```

# Analyze Each Neighborhood

```python
t_onehot=pd.get_dummies(vdf[['VenueCategory']], prefix="", prefix_sep="")
t_onehot['Postal Code']=vdf['Postal Code']
t_onehot['Borough']=vdf['Borough']
t_onehot['Neighborhoods']=vdf['Neighborhood']
fc = list(t_onehot.columns[-3:]) + list(t_onehot.columns[:-3])
t_onehot = t_onehot[fc]
t_onehot
```

#### Next, let's group rows by neighborhood and by taking the mean of the frequency of occurrence of each category

```python
t_onehot1=t_onehot.groupby(["Neighborhoods"]).mean().reset_index()
t_onehot1
```

```python
t_onehot1.shape
```

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

#### Now let's create the new dataframe and display the top 10 venues for each neighborhood.

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhoods']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhoods'] =t_onehot1['Neighborhoods']

for ind in np.arange(t_onehot1.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(t_onehot1.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```

# Cluster Neighborhoods¶

#### Run k-means to cluster the neighborhood into 5 clusters.

```python
kclusters = 5
t_clustering = t_onehot1.drop('Neighborhoods', 1)
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(t_clustering)
kmeans.labels_[0:10] 
```

Let's create a new dataframe that includes the cluster as well as the top 10 venues for each neighborhood.

```python
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)
t_merged = df3.copy()
t_merged = t_merged.join(neighborhoods_venues_sorted.set_index('Neighborhoods'), on='Neighborhood')
t_merged.head()
```

Finally, let's visualize the resulting clusters

```python
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(t_merged['Latitude'], t_merged['Longitude'], t_merged['Neighborhood'], t_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
